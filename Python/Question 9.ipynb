### ● Get all TESLA cars with the model year, and model type made in Bothell City.

### ● Get all the cars that have an electric range of more than 100, and were made after 2015

### ● Draw plots to show the distribution between city and electric vehicle typ

```python
df[['City','Electric Vehicle Type']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting only 100 to avoid congestion of data
plt.figure(figsize=(18, 10))
sns.countplot(data=df[:100], x='City', hue='Electric Vehicle Type')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend(title='Electric Vehicle Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
